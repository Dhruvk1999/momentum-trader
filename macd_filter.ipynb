```python
import pandas as pd 
import numpy as np
import talib
import yfinance as yf
from datetime import date, timedelta

symbols = list(pd.read_csv("niftyfo.csv")["SYMBOL"])
symbols = [x.strip()+".NS" for x in symbols[5:]] 
def macd_filter():
    result = pd.DataFrame()
    start_date = date.today() -timedelta(days = 50)
    end_date = date.today() + timedelta(days=1)
    for symbol in symbols:
        tdata  = yf.download(symbol,start = start_date,end = end_date)
        macd, signal, _ = talib.MACD(tdata['Close'], fastperiod=3, slowperiod=15, signalperiod=9)
        tdata["BuySignal"] = ((macd > 0.95*signal) & (macd.shift(1) <= signal.shift(1))).astype(int)
        tdata["SellSignal"] = ((macd < signal) & (macd.shift(1) >= signal.shift(1))).astype(int)
        
        # Calculating RSI for 4 days and 14 days
        tdata['RSI_4'] = talib.RSI(tdata['Close'], timeperiod=4)
        tdata['RSI_14'] = talib.RSI(tdata['Close'], timeperiod=14)
        
        # Calculating Rate of Change for 3 days
        tdata['ROC_3'] = talib.ROC(tdata['Close'], timeperiod=3)
        
        tdata["Symbol"] = symbol
        
        #appending to the result
        result = pd.concat([result,tdata[tdata.BuySignal == 1]])
        result = pd.concat([result,tdata[tdata.SellSignal == 1]])
    return result
        
        
#     data["BuySignal"] = 1 
#     data["SellSignal"] = 1
#     macd, signal, _ = talib.MACD(data['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
#     data['EMA3'] = data['Close'].ewm(span=3, adjust=False).mean()
#     data['EMA15'] = data['Close'].ewm(span=15, adjust=False).mean()
#     data['BuySignal'] = ((data['EMA3'] > data['EMA15']) & (data['EMA3'].shift(1) <= data['EMA15'].shift(1))).astype(int)
```
