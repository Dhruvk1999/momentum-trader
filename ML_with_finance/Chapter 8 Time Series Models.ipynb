### Topic 2: Decomposing Time Series Patterns

Time series data often contains multiple components, including trends, seasonality, and noise. Decomposing time series patterns involves separating these components to gain a better understanding of the underlying patterns and trends. We'll explore techniques for decomposition and visualization.

#### Time Series Decomposition Methods

1. **Additive Decomposition:**

   The additive decomposition model assumes that a time series can be decomposed into three components: trend, seasonality, and residual (noise). Let's use the `statsmodels` library to perform an additive decomposition and visualize the components.

   ```python
   import statsmodels.api as sm

   # Perform additive decomposition
   decomposition = sm.tsa.seasonal_decompose(data['Close'], model='additive', freq=252)  # Assume daily data (252 trading days)

   # Plot the decomposed components
   plt.figure(figsize=(12, 10))
   plt.subplot(4, 1, 1)
   plt.plot(decomposition.observed, label='Observed', color='blue')
   plt.title('Observed Price')
   plt.grid()

   plt.subplot(4, 1, 2)
   plt.plot(decomposition.trend, label='Trend', color='orange')
   plt.title('Trend Component')
   plt.grid()

   plt.subplot(4, 1, 3)
   plt.plot(decomposition.seasonal, label='Seasonal', color='green')
   plt.title('Seasonal Component')
   plt.grid()

   plt.subplot(4, 1, 4)
   plt.plot(decomposition.resid, label='Residual', color='red')
   plt.title('Residual Component')
   plt.grid()

   plt.tight_layout()
   plt.show()
   ```

   In this code, we assume daily data and perform an additive decomposition of the closing prices into trend, seasonal, and residual components. We then visualize each component separately.

2. **Multiplicative Decomposition:**

   Similar to additive decomposition, the multiplicative decomposition model assumes that a time series can be decomposed into trend, seasonality, and residual components, but with multiplicative relationships. You can perform multiplicative decomposition using `model='multiplicative'` in the code above.

#### Interpretation and Usage

Decomposing time series data helps traders and analysts understand the underlying patterns and make informed decisions. It can aid in trend analysis, identifying seasonality, and separating out irregular variations (residuals) for further analysis.

By decomposing time series data, you can better visualize and model each component separately, which can be useful for forecasting, anomaly detection, and trading strategy development.

Feel free to explore the decomposed components and let me know if you have any questions or if you'd like to proceed to the next topic: "How to Compute Rolling Window Statistics."

````python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import yfinance as yf

data = yf.download("HDFCAMC.NS",start = "2018-06-22")

# Perform additive decomposition
decomposition = sm.tsa.seasonal_decompose(data['Close'], model='multiplicative', period =252)  # Assume daily data (252 trading days)

# Plot the decomposed components
plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.plot(decomposition.observed, label='Observed', color='blue')
plt.title('Observed Price')
plt.grid()

plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.grid()

plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Seasonal', color='green')
plt.title('Seasonal Component')
plt.grid()

plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Residual', color='red')
plt.title('Residual Component')
plt.grid()

plt.tight_layout()
plt.show()
````

### Topic 3: How to Compute Rolling Window Statistics

Rolling window statistics involve calculating summary statistics, such as moving averages, over a fixed-size window that slides or "rolls" through the time series data. These statistics are useful for smoothing out noise and identifying trends. We'll cover how to compute rolling window statistics using Python and `pandas`.

#### Computing Rolling Window Statistics

1. **Simple Moving Average (SMA):**

   The Simple Moving Average is a commonly used rolling window statistic. It calculates the mean (average) of a specified window of data points.

   ```python
   # Calculate a 50-day Simple Moving Average (SMA)
   data['50_SMA'] = data['Close'].rolling(window=50).mean()
   ```

2. **Exponential Moving Average (EMA):**

   The Exponential Moving Average gives more weight to recent data points. It can be calculated using the `ewm` method in `pandas`.

   ```python
   # Calculate a 50-day Exponential Moving Average (EMA)
   data['50_EMA'] = data['Close'].ewm(span=50, adjust=False).mean()
   ```

   Here, we calculate a 50-day EMA of the closing prices.

3. **Other Rolling Window Statistics:**

   You can compute various other rolling window statistics, such as rolling standard deviation, rolling maximum, and rolling minimum, to capture different aspects of the data.

   # Calculate rolling standard deviation (volatility) over a 20-day window
   data['Volatility'] = data['Close'].rolling(window=20).std()

   These calculations provide insights into volatility, short-term highs, and short-term lows in the data.

### Topic 4: How to Measure Autocorrelation

Autocorrelation, also known as serial correlation, is a crucial concept in time series analysis. It measures the correlation between a time series and its lagged values. Autocorrelation helps us understand whether there is a systematic relationship between past and future observations. We'll explore how to measure autocorrelation using Python and `pandas`.

#### How to Measure Autocorrelation

1. **Autocorrelation Function (ACF):**

   The Autocorrelation Function (ACF) measures the correlation between a time series and its lagged values at different lags. It is a fundamental tool for identifying seasonality and patterns in time series data.

   ```python
   from statsmodels.graphics.tsaplots import plot_acf

   # Calculate and plot the ACF
   plot_acf(data['Close'], lags=50)
   plt.title('Autocorrelation Function (ACF)')
   plt.xlabel('Lag')
   plt.ylabel('ACF Value')
   plt.grid()
   plt.show()
   ```

   In this code, we use the `plot_acf` function from `statsmodels` to calculate and plot the ACF for the closing prices of a stock. The `lags` parameter specifies the number of lags to include in the plot.

   The ACF plot helps identify significant autocorrelations at different lags.

2. **Partial Autocorrelation Function (PACF):**

   The Partial Autocorrelation Function (PACF) measures the correlation between a time series and its lagged values while adjusting for the influence of intermediate lags. It is useful for determining the order of autoregressive (AR) models.

   ```python
   from statsmodels.graphics.tsaplots import plot_pacf

   # Calculate and plot the PACF
   plot_pacf(data['Close'], lags=50)
   plt.title('Partial Autocorrelation Function (PACF)')
   plt.xlabel('Lag')
   plt.ylabel('PACF Value')
   plt.grid()
   plt.show()
   ```

   Similar to ACF, we use the `plot_pacf` function to calculate and plot the PACF.

#### Interpretation

- In both the ACF and PACF plots, significant spikes or peaks at specific lags indicate autocorrelation. These spikes suggest that past observations at those lags have a relationship with future values.

- The ACF plot helps identify the order of moving average (MA) models, while the PACF plot is useful for determining the order of autoregressive (AR) models.

- The choice of lag order for ARIMA modeling often depends on the patterns observed in the ACF and PACF plots.

Measuring autocorrelation is essential for understanding the temporal dependencies within time series data and selecting appropriate models for forecasting and analysis.

### Topic 5: How to Diagnose and Achieve Stationarity

Stationarity is an important concept in time series analysis. Stationary time series data has statistical properties, such as mean and variance, that remain constant over time. Non-stationary data may have trends or seasonality, making it challenging to model. In this topic, we'll explore how to diagnose stationarity and achieve it if necessary.

#### Diagnosing Stationarity

1. **Visual Inspection:**

   One way to diagnose stationarity is by visual inspection. Plotting the time series data and looking for trends or patterns can provide initial insights.

   ```python
   # Plot the time series data
   plt.figure(figsize=(12, 6))
   plt.plot(data['Close'], label='Closing Price', color='blue')
   plt.title('Stock Price Over Time')
   plt.xlabel('Date')
   plt.ylabel('Price')
   plt.legend()
   plt.grid()
   plt.show()
   ```

   If the plot shows an obvious trend or seasonality, the data may not be stationary.

2. **Summary Statistics:**

   You can also calculate summary statistics for different time periods to check for stationarity. Calculate the mean and variance for subsets of the data and see if they vary significantly.

   ```python
   # Calculate rolling mean and variance
   rolling_mean = data['Close'].rolling(window=20).mean()
   rolling_var = data['Close'].rolling(window=20).var()

   # Plot rolling statistics
   plt.figure(figsize=(12, 6))
   plt.plot(data['Close'], label='Closing Price', color='blue')
   plt.plot(rolling_mean, label='Rolling Mean', color='orange', linestyle='--')
   plt.plot(rolling_var, label='Rolling Variance', color='green', linestyle='--')
   plt.title('Rolling Mean and Variance')
   plt.xlabel('Date')
   plt.ylabel('Price / Variance')
   plt.legend()
   plt.grid()
   plt.show()
   ```

   If the rolling mean or variance shows a clear trend, it suggests non-stationarity.

#### Achieving Stationarity

1. **Differencing:**

   One common method to achieve stationarity is differencing. Differencing involves subtracting a lagged value from the current value. You can apply differencing multiple times if needed.

   ```python
   # Calculate first-order differences
   data['First_Difference'] = data['Close'].diff()

   # Plot differenced data
   plt.figure(figsize=(12, 6))
   plt.plot(data['First_Difference'], label='First-Order Difference', color='blue')
   plt.title('Differenced Data')
   plt.xlabel('Date')
   plt.ylabel('Price Difference')
   plt.legend()
   plt.grid()
   plt.show()
   ```

   Differenced data may exhibit stationarity, as shown in the plot.

2. **Augmented Dickey-Fuller Test:**

   The Augmented Dickey-Fuller (ADF) test is a statistical test to formally check for stationarity. A p-value less than a certain significance level indicates stationarity.

   ```python
   from statsmodels.tsa.stattools import adfuller

   # Perform the ADF test on the differenced data
   result = adfuller(data['First_Difference'].dropna())

   # Check the p-value
   p_value = result[1]
   if p_value <= 0.05:
       print('Differenced data is stationary (p-value = {:.4f})'.format(p_value))
   else:
       print('Differenced data is non-stationary (p-value = {:.4f})'.format(p_value))
   ```

   The ADF test helps you confirm whether differenced data is stationary.

Achieving stationarity is crucial for time series modeling, as stationary data simplifies the modeling process and makes it easier to apply forecasting techniques. If you have any questions or would like to explore this topic further or move on to the next, please let me know.

Heteroscedasticity, often referred to as heteroedasticity, is a statistical term used to describe a situation in which the variability or dispersion of the dependent variable (i.e., the "errors" or "residuals") in a regression model is not constant across all levels of the independent variable(s). In simpler terms, it means that the spread or variability of the data points is unequal as you move along the range of predictor(s).

Heteroscedasticity is a violation of one of the key assumptions of linear regression models, known as homoscedasticity or constant variance. In linear regression, it is assumed that the variance of the residuals is the same for all values of the independent variables. When this assumption is not met, it can have several implications:

1. **Inefficient Estimates:** Heteroscedasticity can lead to inefficient parameter estimates, which means that the estimates of the coefficients may not be the most precise or reliable.

2. **Biased Standard Errors:** Standard errors of the coefficient estimates may be biased, which affects the accuracy of hypothesis tests and confidence intervals.

3. **Incorrect Inference:** Heteroscedasticity can lead to incorrect statistical inference, such as t-tests and F-tests, which are commonly used to assess the significance of model coefficients and overall model fit.

4. **Misleading Conclusions:** It can also lead to misleading conclusions about the strength and significance of relationships between independent and dependent variables.

5. **Model Fit:** Models that assume constant variance may provide poor fit to heteroscedastic data.

Detecting heteroscedasticity is typically done by examining residual plots, such as scatterplots of residuals against the predicted values or the independent variables. If the spread of residuals widens or narrows as you move along the range of predicted values, it's an indication of heteroscedasticity.

Addressing heteroscedasticity often involves transforming the dependent variable, using weighted least squares regression, or selecting appropriate modeling techniques that are robust to heteroscedasticity, such as generalized least squares (GLS) or robust regression.

In the context of time series analysis, heteroscedasticity can be particularly relevant when dealing with financial or economic data, as financial time series often exhibit changing volatility over time. Identifying and addressing heteroscedasticity is crucial for accurate modeling and reliable statistical inference.

Autoregressive (AR) and Moving Average (MA) models, while similar to linear regression models in some aspects, are fundamentally different in their formulation and use cases:

**Autoregressive (AR) Models:**

1. **Formulation:** AR models capture the relationship between a time series and its past values. Mathematically, an AR(p) model is defined as:
$$
   \[Y_t = c + \phi_1Y_{t-1} + \phi_2Y_{t-2} + \ldots + \phi_pY_{t-p} + \varepsilon_t\]
$$
   Here, \(Y_t\) is the current value of the time series, \(c\) is a constant, \(\phi_1, \phi_2, \ldots, \phi_p\) are the autoregressive coefficients, \(Y_{t-1}, Y_{t-2}, \ldots, Y_{t-p}\) are the lagged values of the time series, and \(\varepsilon_t\) is a white noise error term.

2. **Use Case:** AR models are used to model and forecast time series data where the current value depends on its past values. They are suitable for capturing serial correlation and temporal dependencies within a time series.

**Moving Average (MA) Models:**

1. **Formulation:** MA models focus on modeling the relationship between a time series and its past prediction errors (residuals). Mathematically, an MA(q) model is defined as:

$$   \[Y_t = \mu + \varepsilon_t + \theta_1\varepsilon_{t-1} + \theta_2\varepsilon_{t-2} + \ldots + \theta_q\varepsilon_{t-q}\]$$

   Here, \(Y_t\) is the current value of the time series, \(\mu\) is the mean of the time series, \(\varepsilon_t\) is the white noise error term at time \(t\), and \(\theta_1, \theta_2, \ldots, \theta_q\) are the moving average coefficients applied to past error terms.

2. **Use Case:** MA models are used to model and forecast time series data where the current value depends on past prediction errors. They are often used in combination with autoregressive models to create Autoregressive Moving Average (ARMA) models.

**Linear Regression Models:**

1. **Formulation:** Linear regression models, on the other hand, are used to model the relationship between a dependent variable and one or more independent variables. The formulation is typically linear, and it aims to estimate the coefficients that best describe this linear relationship.

$$   \[Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n + \varepsilon\]$$

   Here, \(Y\) is the dependent variable, \(X_1, X_2, \ldots, X_n\) are the independent variables, \(\beta_0, \beta_1, \beta_2, \ldots, \beta_n\) are the coefficients, and \(\varepsilon\) is the error term.

2. **Use Case:** Linear regression models are used for various purposes, including prediction, inference, and understanding the relationships between variables. They are not specialized for time series data modeling but are applicable to a wide range of data types.

In summary, AR and MA models are designed specifically for modeling time series data, while linear regression models are more general and can be applied to various types of data. While there are similarities in terms of estimating coefficients, the underlying assumptions and goals of these models are different. Therefore, the choice between them depends on the nature of the data and the specific modeling objectives.

### Topic 8: How to Build Autoregressive Models

Autoregressive (AR) models are a class of univariate time series models that capture the relationship between a time series and its past values. AR models are useful for understanding and forecasting time series data that exhibit autocorrelation, where values at one time point are correlated with previous values. In this topic, we'll explore how to build autoregressive models using Python.

#### Autoregressive (AR) Models

An autoregressive model of order \(p\) is denoted as AR(p). The general form of an AR(p) model is:

\[Y_t = c + \phi_1Y_{t-1} + \phi_2Y_{t-2} + \ldots + \phi_pY_{t-p} + \varepsilon_t\]

Here's how to build an AR(p) model:

1. **Data Preparation:** Ensure that your time series data is stationary, as AR models assume stationarity. If your data is not stationary, consider differencing or other transformations as discussed earlier.

### Autoregressive Conditional Heteroskedasticity (ARCH) Model

The ARCH model, introduced by Robert Engle, assumes that the variance of the time series follows an autoregressive process based on past squared returns. The ARCH(p) model can be expressed as:

\[
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{p} \alpha_i r_{t-i}^2
\]

Where:
- \(\sigma_t^2\) is the conditional variance at time \(t\).
- \(\alpha_0\) is a constant.
- \(\alpha_i\) are parameters to be estimated.
- \(r_{t-i}\) are the squared returns at time \(t-i\).

ARCH models are used to capture volatility clustering, which means that periods of high volatility are followed by other high-volatility periods.

### Generalized Autoregressive Conditional Heteroskedasticity (GARCH) Model

The GARCH model, introduced by Tim Bollerslev, is an extension of the ARCH model that includes lagged conditional variances in the model. The GARCH(p, q) model can be expressed as:

\[
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{p} \alpha_i r_{t-i}^2 + \sum_{j=1}^{q} \beta_j \sigma_{t-j}^2
\]

Where:
- \(\sigma_t^2\) is the conditional variance at time \(t\).
- \(\alpha_0\) is a constant.
- \(\alpha_i\) and \(\beta_j\) are parameters to be estimated.
- \(r_{t-i}\) are the squared returns at time \(t-i\).
- \(\sigma_{t-j}^2\) are the conditional variances at time \(t-j\).

GARCH models are more flexible than ARCH models because they allow the modeling of volatility persistence, which means that past volatility affects future volatility.

### Topic 14: Generalizing ARCH  the GARCH Model

The Generalized Autoregressive Conditional Heteroskedasticity (GARCH) model is an extension of the Autoregressive Conditional Heteroskedasticity (ARCH) model. #### Key Concepts in the GARCH Model

The GARCH model builds upon the key concepts of the ARCH model and introduces a few additional concepts:

1. **Heteroskedasticity:** Like the ARCH model, the GARCH model deals with heteroskedasticity, which is the time-varying variance observed in financial time series data.

2. **Conditional Variance:** The GARCH model defines the conditional variance, denoted as \(\sigma_t^2\), as the variance of the time series at time \(t\) given past information.

3. **Lagged Squared Returns:** Similar to the ARCH model, the GARCH model assumes that the conditional variance \(\sigma_t^2\) depends on past squared returns \(r_{t-1}^2, r_{t-2}^2, \ldots, r_{t-p}^2\) (where \(p\) is the order of the model).

5. **Parameter Estimation:** The GARCH model estimates parameters (\(\alpha_0, \alpha_1, \alpha_2, \ldots, \beta_1, \beta_2, \ldots\)) that determine how past squared returns and past conditional variances influence the current conditional variance.

#### GARCH(p, q) Model

\[
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{p} \alpha_i r_{t-i}^2 + \sum_{j=1}^{q} \beta_j \sigma_{t-j}^2
\]

Where:

### Topic 17: Cointegration - Time Series with a Common Trend

Cointegration is a critical concept in multivariate time series analysis, particularly in financial data. It deals with the idea that even though multiple time series variables may individually appear non-stationary (i.e., they have trends), there can exist a linear combination of these variables that is stationary. In simpler terms, cointegrated time series share a common long-term trend, allowing them to be analyzed together.

#### Key Concepts in Cointegration

Here are some key concepts related to cointegration:

1. **Non-stationarity:** A non-stationary time series is one that exhibits a trend or has a changing mean and variance over time. Non-stationary series are challenging to analyze because their statistical properties change over time.

2. **Stationarity:** A stationary time series has a constant mean and variance over time. Stationary series are easier to model and analyze.

3. **Cointegration:** Cointegration is a statistical property that exists when two or more non-stationary time series variables have a long-term relationship. In other words, a linear combination of these variables is stationary.

4. **Engle-Granger Cointegration Test:** This is a common statistical test used to determine whether two variables are cointegrated. If they are, it means they share a common long-term trend.

5. **Cointegrating Vector:** In a cointegrated system, there is a linear combination (cointegrating vector) of the variables that results in a stationary time series. This vector represents the long-term relationship between the variables.

#### Why Cointegration Matters in Finance

Cointegration is particularly important in finance for several reasons:

- **Pairs Trading:** Cointegrated assets can be used in pairs trading strategies. When two cointegrated assets deviate from their long-term relationship (spread), traders can take positions to profit from mean reversion.

- **Risk Management:** Understanding cointegrated assets can help in risk management and portfolio optimization. It allows for better diversification when assets are not perfectly correlated but share a common trend.

- **Macroeconomic Analysis:** Cointegration is used to analyze macroeconomic indicators that are expected to move together over the long term, such as interest rates and inflation.

#### Python Code Example for Cointegration Test

Here's a Python code example using the Engle-Granger cointegration test to check if two time series variables are cointegrated:

```python
import pandas as pd
from statsmodels.tsa.stattools import coint

# Load your time series data
data = pd.read_csv('your_data.csv', index_col='Date', parse_dates=True)

series2 = data['Series2']

print("Cointegration test p-value:", result[1])

# Interpret the result
if result[1] < 0.05:
    print("The two series are likely cointegrated.")
else:
    print("The two series are likely not cointegrated.")
```

This code conducts the Engle-Granger cointegration test between two time series variables and interprets the test result based on a significance level (e.g., 0.05).

Understanding cointegration is crucial when working with multiple financial time series variables, as it enables better modeling, trading, and risk management strategies.

### Topic 18: How to Use Cointegration for Pairs Trading Strategy

Pairs trading is a popular trading strategy in financial markets, and it relies on the concept of cointegration. The strategy involves identifying two or more assets that are cointegrated, meaning they share a common long-term relationship. Traders then take advantage of short-term deviations from this relationship to generate trading signals.

#### Key Concepts in Pairs Trading

Here are some key concepts related to pairs trading using cointegration:

1. **Cointegrated Pairs:** In pairs trading, you look for pairs of assets (e.g., stocks) that are cointegrated. This means they have a common long-term trend.

2. **Spread:** The spread is the difference between the prices (or returns) of the two assets in a cointegrated pair. When the spread deviates from its historical average, it presents a trading opportunity.

3. **Mean Reversion:** The pairs trading strategy assumes that the spread will revert to its historical mean over time. Traders take positions when they expect the spread to narrow or widen.

4. **Entry and Exit Points:** Traders establish long and short positions in the assets of a cointegrated pair. Entry points are often based on deviation from the mean spread, while exit points are based on reversion to the mean.

5. **Risk Management:** Proper risk management is essential in pairs trading. This includes setting stop-loss levels and position sizing to limit potential losses.

#### Steps to Implement a Pairs Trading Strategy

Here are the general steps to implement a pairs trading strategy using cointegration:

1. **Identify Cointegrated Pairs:** Use statistical tests like the Engle-Granger test or Johansen test to identify cointegrated pairs from a universe of assets.

2. **Calculate the Spread:** Calculate the spread between the prices (or returns) of the assets in the cointegrated pair.

3. **Determine Entry and Exit Criteria:** Set criteria for entering and exiting trades based on deviations from the mean spread or other technical indicators.

4. **Risk Management:** Implement risk management measures, including setting stop-loss levels and position sizing.

5. **Backtesting:** Backtest your trading strategy using historical data to assess its performance.

6. **Live Trading:** If the backtesting results are satisfactory, implement the strategy in live trading.

#### Python Code Example for Pairs Trading

Here's a simplified Python code example for pairs trading using cointegration:

```python
import pandas as pd
from statsmodels.tsa.stattools import coint
import numpy as np

# Load historical price data for two cointegrated assets
data = pd.read_csv('your_data.csv', index_col='Date', parse_dates=True)
asset1 = data['Asset1']
asset2 = data['Asset2']

# Perform the cointegration test
result = coint(asset1, asset2)

# Check if the assets are cointegrated (using a significance level of 0.05)
if result[1] < 0.05:
    print("The two assets are cointegrated.")

    # Calculate the spread (difference) between the assets
    spread = asset1 - result[0] * asset2
    
    # Define entry and exit criteria based on spread deviations or other indicators
    entry_threshold = 2.0 * np.std(spread)
    exit_threshold = 0.5 * np.std(spread)
    
    # Implement a simple pairs trading strategy
    position = 0  # 0: No position, 1: Long asset1, -1: Short asset1
    trades = []
    
    for i in range(1, len(data)):
        if spread[i] > entry_threshold and position != 1:
            # Go long on asset1
            trades.append((data.index[i], "Long Asset1"))
            position = 1
        elif spread[i] < -entry_threshold and position != -1:
            # Go short on asset1
            trades.append((data.index[i], "Short Asset1"))
            position = -1
        elif abs(spread[i]) < exit_threshold and position != 0:
            # Exit the position
            trades.append((data.index[i], "Exit Position"))
            position = 0
    
    # Print the trading signals
    for trade in trades:
        print(trade)

else:
    print("The two assets are not cointegrated.")
```

This code performs a cointegration test between two assets, calculates the spread, and implements a basic pairs trading strategy based on spread deviations.

Please note that this is a simplified example, and real-world trading strategies involve more complexities, including transaction costs, slippage, and continuous monitoring. Additionally, proper risk management is essential when implementing trading strategies.
